## Handling outlier of Fare , Parch and SibSp columns

```python
column_name = ['SibSp','Parch','Fare']
```

```python
Q1 = df[column_name].quantile(0.25)
Q3 = df[column_name].quantile(0.75)

# Calculate the IQR (Interquartile Range)
IQR = Q3 - Q1

# Define the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
df[column_name] = (df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)
```

```python
numeric_columns = df.select_dtypes(include=['int', 'float']).columns

num_cols = len(numeric_columns)

fig, axes = plt.subplots(nrows=num_cols, ncols=1, figsize=(8, 3 * num_cols))

if num_cols == 1:
    axes = [axes]

for ax, col in zip(axes, numeric_columns):
    sns.boxplot(data=df, x=col, ax=ax)
    ax.set_title(f'Boxplot of {col}')
    ax.set_xlabel('')

plt.tight_layout()
plt.show()
```
